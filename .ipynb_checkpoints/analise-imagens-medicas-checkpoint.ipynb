## 4.5 Indicadores de desempenho

O Keras não possui os indicadores de desempenho como precisão, sensibilidade e pontuação f1 por padrão, portanto precisamos implementar externamente.

````python
def recall_score(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    recall = true_positives / (possible_positives + K.epsilon())
    return recall

def precision_score(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    return precision

def f1_score(y_true, y_pred):
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+K.epsilon()))
````

A ResNet introduziu o primeiro conceito de skip connection, a imagem da esquerda é uma rede convolucional normal, ela esta "empilhando" as camadas de convolução, uma após a outra, já a imagem da direita é uma rede residual, ainda existe o mesmo empilhamento, mas agora também é adicionado a entrada original e saída do bloco, isso é o skip connection, a ideia aqui é que a rede profunda aprenda igual uma rede mais rasa, copiando as ativações da camada mais rasa.

Essa imagem abaixo mostra uma comparação entre uma rede VGG-19, outra rede com 34-layer plana (normal) e uma rede 34-layer residual (usando o metodo que falei acima), fica claro qual foi a ideia que eles tiveram para resolver o problema em redes profundas

Usando essa tecnica em teoria pode-se construir redes de qualquer profundidade que novas camadas estão realmente ajudando a aprender novos padrões subjacentes nos dados de entrada.

A ResNet50 é uma variação da ResNet, como ela existe outras tantas variações dela, como a ResNet152 e ResNet101, ela se constituiu de 5 estágios, cada um com uma convolução e um bloco de identidade, cada bloco de convolução possuiu 3 camadas de convolução e cada bloco de identidade possuiu 3 camadas de convolução, a ResNet50 possuiu mais de 23 milhões de parametros treináveis.

A técnica de transfer learning consiste de utilizar o mesmo modelo e treiná-lo para outas imagens. Por tal motivo, excluímos a última camada para modelar com as classes que definimos, ou seja, **controle**, **bacteriana** e **viral**.

Informe a quantidade de classes a serem classificadas.

````python
## IMPLEMENTE

qtde_classes = 3
````

````python
conv_base = ResNet50(include_top=False)

for layer in conv_base.layers:
    layer.trainable = False
````

````python
x = conv_base.output
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(128, activation='relu')(x) 

predictions = layers.Dense(qtde_classes, activation='softmax')(x)
model = Model(conv_base.input, predictions)

model.summary()
````

````python
optimizer = keras.optimizers.Adam()
model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=[precision_score, recall_score, f1_score])
````

O número de épocas define quantas vezes o modelo irá treinar e validar o erro, assim ajustando os pesos para melhor convergência.
Escolha o número adequado de épocas para alcançarmos pelo menos **70% de precisão de validação**.

````python
## IMPLEMENTE

qtde_epocas = 100
````

````python
history = model.fit_generator(generator=train_generator, 
                              epochs=qtde_epocas, 
                              validation_steps=5, 
                              steps_per_epoch=5, 
                              validation_data=val_generator)
````

Um modelo que converge bem possui o gráfico de perda (*loss*) descendente e os gráfico de precisão (*precision*), sensibilidade (*recall*) e pontuação f1 (*f1 score*) em acendente.

````python
# Exibindo dados de Precisão
plt.plot(history.history['precision_score'])
plt.plot(history.history['val_precision_score'])
plt.title('model precision')
plt.ylabel('precision')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# Exibindo dados de Sensibilidade
plt.plot(history.history['recall_score'])
plt.plot(history.history['val_recall_score'])
plt.title('model recall')
plt.ylabel('recall')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# Exibindo dados de F1 Score
plt.plot(history.history['f1_score'])
plt.plot(history.history['val_f1_score'])
plt.title('model f1_score')
plt.ylabel('f1_score')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# Exibindo dados de Perda
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
````

### 4.5.2 Arquitetura VGG16

**Pergunta**: Explique como é constituída a arquitetura do VGG16? *Utilize, se necessário, gráficos, projetos que utilizam essa arquitetura. Detalhe também sua topologia em camadas e mostre quais as situações essa arquitetura pode ter mais êxito e quais cenários não tem.*

**Resposta**:

Trata-se de uma rede CNN no qual possui as seguintes características:
•	Contém um total de 16 layers, onde os pesos e parâmetros de tendências são aprendidos. 
•	Um total de 13 camadas convolucionais são empilhadas uma após a outra e 3 camadas densas para classificação.
•	O número de filtros nas camadas de convolução segue um padrão crescente
•	Os recursos informativos são obtidos por camadas Max de pooling aplicadas em diferentes etapas da arquitetura.
•	As camadas densas são compostas por 4096, 4096 e 1000 nós cada.
•	Os contras dessa arquitetura são que ela demora para treinar e produz o modelo com um tamanho muito grande.
O bloco de construção mais importante de uma CNN é a camada convolucional: 6 neurônios em a primeira camada convolucional não está conectada a todos os pixels da imagem de entrada, mas apenas para pixels em seus campos receptivos (exemplificado na figura abaixo). Por sua vez, cada neurônio na segunda camada convolucional é conectado apenas para neurônios localizados dentro de um pequeno retângulo na primeira camada. Esta arquitetura permite que a rede se concentre em pequenos recursos de baixo nível na primeira camada oculta, em seguida, monte-os em recursos maiores de nível superior na próxima camada oculta, e assim sucessivamente. Essa estrutura hierárquica é comum em imagens do mundo real, que é uma das razões pelas quais as CNNs funcionam tão bem para reconhecimento de imagem.

<p align="center">
    <img src="imagens/figura1.jpg">
</p>
Figura 1 - Camadas CNN com campos receptores locais retangulares

Um neurônio localizado na linha i, coluna j de uma determinada camada está conectado às saídas do neurônios na camada anterior localizados nas linhas i para i + fh - 1, colunas j para j + fw - 1, onde fh e fw são a altura e a largura do campo receptivo (veja a Figura 2). Para que uma camada tenha a mesma altura e largura da camada anterior, é comum adicionar zeros ao redor das entradas, conforme mostrado no diagrama. Isso é chamado de preenchimento de zero.

<p align="center">
    <img src="imagens/figura2.jpg">
</p>
 
Figura 2 - Conexões entre camadas e preenchimento zero.

Também é possível conectar uma grande camada de entrada a uma camada muito menor espaçando os campos receptivos, conforme mostrado na Figura 14-4. A mudança de um campo receptivo para o a seguir é chamada de passada. No diagrama, uma camada de entrada 5 × 7 (mais preenchimento de zero) está conectada para uma camada 3 × 4, usando campos receptivos 3 × 3 e uma passada de 2 (neste exemplo a passada é a mesma em ambas as direções, mas não tem que ser assim). Um neurônio localizado na linha i, a coluna j na camada superior está conectada às saídas dos neurônios em a camada anterior localizada nas linhas i × sh a i × sh + fh - 1, colunas j × sw a j × sw + fw -1, onde sh e sw são as passadas verticais e horizontais.

<p align="center">
    <img src="imagens/figura3.jpg">
</p>

Figura 3 - Reduzindo a dimensionalidade usando uma passada dois.

Os pesos de um neurônio podem ser representados como uma pequena imagem do tamanho do campo receptivo. Por exemplo, a Figura 4 mostra dois conjuntos possíveis de pesos, chamados filtros (ou convolução kernels). O primeiro é representado como um quadrado preto com uma linha branca vertical em o meio (é uma matriz 7 × 7 cheia de 0s, exceto para a coluna central, que está cheia de 1s); neurônios usando esses pesos irão ignorar tudo em seu campo receptivo, exceto para a linha vertical central (uma vez que todas as entradas serão multiplicadas por 0, exceto para o aqueles localizados na linha vertical central). 
O segundo filtro é um quadrado preto com uma linha branca horizontal no meio. Mais uma vez, os neurônios que usam esses pesos vão ignore tudo em seu campo receptivo, exceto para a linha horizontal central. Agora, se todos os neurônios em uma camada usam o mesmo filtro de linha vertical (e o mesmo termo de polarização), e você alimenta a rede com a imagem de entrada mostrada na Figura 4 (imagem inferior), a camada produzirá a imagem superior esquerda. Observe que as linhas brancas verticais aumentam enquanto o resto fica borrado. Da mesma forma, a imagem superior direita é o que você obtém se todos os neurônios usam o mesmo filtro de linha horizontal; observe que as linhas brancas horizontais aprimorado enquanto o resto é desfocado. Assim, uma camada cheia de neurônios usando o mesmo o filtro produz um mapa de características, que destaca as áreas em uma imagem que ativam o filtrar mais. Claro que você não precisa definir os filtros manualmente: em vez disso, durante treinar a camada convolucional aprenderá automaticamente os filtros mais úteis para sua tarefa, e as camadas acima aprenderão a combiná-los em padrões mais complexos.


<p align="center">
    <img src="imagens/figura4.jpg">
</p>
Figura 4 - Aplicando dois filtros diferentes para obter dois mapas de recursos.

Até agora, a saída de cada camada convolucional foi representada como uma camada 2D fina, mas na realidade uma camada convolucional tem vários filtros (você decide quantos), e gera um mapa de recursos por filtro, para que seja representado com mais precisão em 3D (consulte a Figura 14-6). Para fazer isso, ele tem um neurônio por pixel em cada recurso mapa, e todos os neurônios dentro de um determinado mapa de recursos compartilham os mesmos parâmetros (ou seja, o mesmos pesos e termo de viés). No entanto, os neurônios em diferentes mapas de recursos usam diferentes parâmetros. O campo receptivo de um neurônio é o mesmo descrito anteriormente, mas estende-se por todos os mapas de recursos das camadas anteriores. Em suma, uma camada convolucional aplica simultaneamente vários filtros treináveis às suas entradas, tornando-o capaz de detectar vários recursos em qualquer lugar em suas entradas.
Além disso, as imagens de entrada também são compostas por várias subcamadas: uma por canal de cor. Normalmente existem três: vermelho, verde e azul (RGB). As imagens em tons de cinza têm apenas um canal, mas algumas imagens podem ter muito mais - por exemplo, imagens de satélite que capturam frequências extras de luz (como infravermelho).

<p align="center">
    <img src="imagens/figura5.jpg">
</p>
 
Figura 5 - Camadas de convolução com vários mapas de recursos e imagens com três cores canais

Especificamente, um neurônio localizado na linha i, coluna j do mapa de características k em uma determinada camada convolucional l está conectado às saídas dos neurônios na camada anterior l - 1, localizada nas linhas i × sh a i × sh + fh - 1 e colunas j × sw a j × sw + fw - 1, em todos os mapas de feições (na camada l - 1). Observe que todos os neurônios localizados na mesma linha i e coluna j, mas em mapas de características diferentes, estão conectados às saídas dos mesmos neurônios na camada anterior.
